# 01 - The Met API Pilot Review

Este notebook revisa la extraccion inicial de textiles andinos desde The Metropolitan Museum of Art Collection API.

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from PIL import Image
```

```python
metadata_path = Path("../data/processed/corpus_met_textiles_andinos_v1_inventario_base.csv")

df = pd.read_csv(metadata_path)

print("Registros:", len(df))
df.head()
```

```python
df.columns.tolist()
```

```python
quality_summary = pd.DataFrame({
    "column": df.columns,
    "non_null": [df[col].notna().sum() for col in df.columns],
    "nulls": [df[col].isna().sum() for col in df.columns],
    "unique_values": [df[col].nunique() for col in df.columns],
})

quality_summary.sort_values("non_null", ascending=False)
```

```python
image_dir = Path("../outputs/samples/met")
image_paths = sorted(image_dir.glob("*.jpg"))

print("Imágenes disponibles:", len(image_paths))
image_paths[:5]
```

```python
n = min(12, len(image_paths))

plt.figure(figsize=(12, 8))

for i, image_path in enumerate(image_paths[:n]):
    img = Image.open(image_path)
    plt.subplot(3, 4, i + 1)
    plt.imshow(img)
    plt.axis("off")
    plt.title(image_path.stem, fontsize=8)

plt.tight_layout()
plt.show()
```
